```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from keras.utils import to_categorical
from keras.optimizers import Adam
from sklearn.metrics import confusion_matrix, classification_report
from typing import Tuple, List, Dict
from dataclasses import dataclass

@dataclass
class Config:
    batch_size: int = 128
    epochs: int = 10
    num_classes: int = 10
    input_shape: Tuple[int, int, int] = (28, 28, 1)
    learning_rate: float = 0.001

class DataHandler:
    def __init__(self, config: Config):
        self.config = config

    def load_and_preprocess(self) -> Tuple[np.ndarray, np.ndarray, np.ndarray, np.ndarray]:
        try:
            (X_train, y_train), (X_test, y_test) = mnist.load_data()

            # Normalize and reshape
            X_train = X_train.astype('float32') / 255.0
            X_test = X_test.astype('float32') / 255.0
            X_train = X_train.reshape(-1, *self.config.input_shape)
            X_test = X_test.reshape(-1, *self.config.input_shape)

            # Convert labels to categorical
            y_train_cat = to_categorical(y_train, self.config.num_classes)
            y_test_cat = to_categorical(y_test, self.config.num_classes)

            return X_train, y_train, X_test, y_test
        except Exception as e:
            raise RuntimeError(f"Failed to load and preprocess MNIST dataset: {str(e)}")

    def get_class_distribution(self, y_train: np.ndarray) -> pd.Series:
        return pd.Series(y_train).value_counts().sort_index()

class CNNModel:
    def __init__(self, config: Config):
        self.config = config
        self.model = self._build_model()

    def _build_model(self) -> Sequential:
        model = Sequential([
            Conv2D(32, (5, 5), activation='relu', padding='same',
                   input_shape=self.config.input_shape, name='conv1'),
            MaxPooling2D((2, 2), name='pool1'),
            Conv2D(64, (5, 5), activation='relu', padding='same', name='conv2'),
            MaxPooling2D((2, 2), name='pool2'),
            Conv2D(64, (5, 5), activation='relu', padding='same', name='conv3'),
            MaxPooling2D((2, 2), name='pool3'),
            Flatten(),
            Dense(100, activation='relu', name='fc4'),
            Dropout(0.5),
            Dense(self.config.num_classes, activation='softmax', name='fc5')
        ])

        model.compile(
            optimizer=Adam(learning_rate=self.config.learning_rate),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        return model

    def summary(self):
        self.model.summary()

class Trainer:
    def __init__(self, model: Sequential, config: Config):
        self.model = model
        self.config = config

    def train(self, X_train: np.ndarray, y_train: np.ndarray,
              X_test: np.ndarray, y_test: np.ndarray) -> Dict:
        history = self.model.fit(
            X_train, y_train,
            batch_size=self.config.batch_size,
            epochs=self.config.epochs,
            validation_data=(X_test, y_test),
            verbose=1
        )
        return history.history

    def evaluate(self, X_test: np.ndarray, y_test: np.ndarray) -> Tuple[float, float]:
        return self.model.evaluate(X_test, y_test, verbose=0)

    def predict(self, X_test: np.ndarray) -> np.ndarray:
        return np.argmax(self.model.predict(X_test), axis=1)

    def get_test_samples(self, X_test: np.ndarray, y_test: np.ndarray,
                        predictions: np.ndarray, max_samples: int = 12) -> Tuple[List, List, List]:
        samples, true_labels, pred_labels = [], [], []

        for i in range(min(max_samples, len(X_test))):
            samples.append(X_test[i])
            true_labels.append(y_test[i])
            pred_labels.append(predictions[i])

        return samples, true_labels, pred_labels

class Visualization:
    @staticmethod
    def plot_class_distribution(class_distribution: pd.Series):
        plt.figure(figsize=(10, 6))
        plt.subplot(1, 2, 1)
        class_distribution.plot(kind='bar')
        plt.title('Class Distribution in Training Set')
        plt.xlabel('Digit Class')
        plt.ylabel('Number of Samples')
        plt.xticks(rotation=0)

    @staticmethod
    def plot_sample_images(X_train: np.ndarray, y_train: np.ndarray):
        plt.subplot(1, 2, 2)
        fig, axes = plt.subplots(2, 5, figsize=(12, 6))
        fig.suptitle('Sample MNIST Images')

        for i in range(10):
            row, col = i // 5, i % 5
            idx = np.where(y_train == i)[0][0]
            axes[row, col].imshow(X_train[idx].reshape(28, 28), cmap='gray')
            axes[row, col].set_title(f'Digit: {i}')
            axes[row, col].axis('off')

        plt.tight_layout()
        plt.show()

    @staticmethod
    def plot_training_history(history: Dict, epochs: int):
        plt.figure(figsize=(12, 4))

        plt.subplot(1, 2, 1)
        plt.plot(history['accuracy'], label='Training Accuracy')
        plt.plot(history['val_accuracy'], label='Validation Accuracy')
        plt.title('Model Accuracy')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(history['loss'], label='Training Loss')
        plt.plot(history['val_loss'], label='Validation Loss')
        plt.title('Model Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()

        plt.tight_layout()
        plt.show()

    @staticmethod
    def plot_test_results(samples: List, true_labels: List, pred_labels: List):
        plt.figure(figsize=(12, 8))
        for i in range(len(samples)):
            plt.subplot(3, 4, i + 1)
            plt.imshow(samples[i].reshape(28, 28), cmap='gray')
            plt.title(f'True: {true_labels[i]}, Pred: {pred_labels[i]}')
            plt.axis('off')
        plt.suptitle('Test Results - True vs Predicted Labels')
        plt.tight_layout()
        plt.show()

    @staticmethod
    def plot_confusion_matrix(true_labels: np.ndarray, pred_labels: np.ndarray):
        cm = confusion_matrix(true_labels, pred_labels)
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                   xticklabels=range(10), yticklabels=range(10))
        plt.title('Confusion Matrix')
        plt.xlabel('Predicted Label')
        plt.ylabel('True Label')
        plt.show()

def main():
    config = Config()

    # Data handling
    print("Loading MNIST dataset and visualization")
    print("=" * 50)
    data_handler = DataHandler(config)
    X_train, y_train, X_test, y_test = data_handler.load_and_preprocess()
    y_test_cat = to_categorical(y_test, config.num_classes)


    print(f"Training data shape: {X_train.shape}")
    print(f"Training labels shape: {y_train.shape}")
    print(f"Test data shape: {X_test.shape}")
    print(f"Test labels shape: {y_test.shape}")

    class_distribution = data_handler.get_class_distribution(y_train)
    print(f"\nClass distribution in training set:\n{class_distribution}")

    Visualization.plot_class_distribution(class_distribution)
    Visualization.plot_sample_images(X_train, y_train)

    # Model setup
    print("\nModel designing")
    print("=" * 50)
    model = CNNModel(config)
    model.summary()

    # Training
    print("\nTraining the model")
    print("=" * 50)
    trainer = Trainer(model.model, config)
    history = trainer.train(X_train, to_categorical(y_train, config.num_classes),
                          X_test, y_test_cat)

    Visualization.plot_training_history(history, config.epochs)

    # Testing
    print("\nTesting the model")
    print("=" * 50)
    test_loss, test_accuracy = trainer.evaluate(X_test, y_test_cat)
    print(f"Test Loss: {test_loss:.4f}")
    print(f"Test Accuracy: {test_accuracy:.4f}")

    # Predictions and visualization
    predictions = trainer.predict(X_test)
    samples, true_labels, pred_labels = trainer.get_test_samples(X_test, y_test, predictions)
    Visualization.plot_test_results(samples, true_labels, pred_labels)

    Visualization.plot_confusion_matrix(y_test, predictions)

    print("\nClassification Report:")
    print(classification_report(y_test, predictions))

    print("\nKeras CNN Implementation Complete!")
    print("=" * 50)

if __name__ == "__main__":
    main()
```
